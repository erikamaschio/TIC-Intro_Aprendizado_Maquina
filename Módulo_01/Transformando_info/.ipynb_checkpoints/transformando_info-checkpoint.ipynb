```python
import pandas as pd
```

```python
titanic = pd.read_csv('../../data/aulas/titanic.csv')
```

```python
titanic.head()
```

```python
# Exibe a coluna 'sex' do DataFrame 'titanic'
titanic.sex  # Retorna todos os valores da coluna 'sex', que contém informações sobre o sexo dos passageiros
```

```python
# Converte a coluna 'sex' do DataFrame 'titanic' em variáveis dummy (0 e 1) para cada categoria presente
pd.get_dummies(titanic.sex)  # Retorna um DataFrame com colunas separadas para cada categoria ('male', 'female') com valores binários
```

```python
# Cria um DataFrame de variáveis dummy a partir da coluna 'sex' do DataFrame 'titanic'
sex = pd.get_dummies(titanic.sex)  # Armazena as variáveis dummy em uma nova variável chamada 'sex'
```

```python
# Adiciona as colunas 'female' e 'male' ao DataFrame 'titanic', usando os valores gerados em 'sex'
titanic[['female', 'male']] = sex  # Copia as variáveis dummy da variável 'sex' para as novas colunas no DataFrame 'titanic'
```

```python
titanic.head()
```

```python
# Remove a coluna 'sex' do DataFrame 'titanic'
titanic.drop(['sex'], axis=1, inplace=True)  # O parâmetro 'axis=1' indica que a operação deve ser feita nas colunas, e 'inplace=True' altera o DataFrame original diretamente
```

```python
titanic.head()
```

```python
# Remove a coluna 'male' do DataFrame 'titanic'
titanic.drop(['male'], axis=1, inplace=True)  # O parâmetro 'axis=1' indica que a operação deve ser feita nas colunas, e 'inplace=True' altera o DataFrame original diretamente
```

```python
titanic.head()
```

```python
# Filtra as linhas do DataFrame 'titanic' com base em duas condições:
# 1. Onde 'female' é 0 (masculino) e 'who' é 'woman' (inconsistência).
# 2. Onde 'female' é 1 (feminino) e 'who' é 'man' (inconsistência).

titanic.loc[(titanic.female == 0) & (titanic.who == 'woman') | (titanic.female == 1) & (titanic.who == 'man')]
# Retorna as linhas do DataFrame que atendem a uma das condições definidas.
```

```python
# Filtra as linhas do DataFrame 'titanic' com base em duas condições de inconsistência entre as colunas 'survived' e 'alive':
# 1. Onde 'survived' é 0 (não sobreviveu) e 'alive' é 'yes' (contradição).
# 2. Onde 'survived' é 1 (sobreviveu) e 'alive' é 'no' (contradição).

titanic.loc[(titanic.survived == 0) & (titanic.alive == 'yes') | (titanic.survived == 1) & (titanic.alive == 'no')]
# Retorna as linhas do DataFrame que atendem a uma das condições inconsistentes.
```

```python
# Remove as colunas 'who' e 'alive' do DataFrame 'titanic'
titanic.drop(['who', 'alive'], axis=1, inplace=True)  # O parâmetro 'axis=1' indica que as colunas serão removidas, e 'inplace=True' aplica a mudança diretamente no DataFrame original
```

```python
titanic
```

```python
# Converte a coluna 'embark_town' do DataFrame 'titanic' em variáveis dummy (0 e 1) para cada categoria presente
pd.get_dummies(titanic.embark_town)  # Retorna um DataFrame com colunas separadas para cada cidade de embarque ('Southampton', 'Cherbourg', 'Queenstown') com valores binários
```

```python
# Cria variáveis dummy a partir da coluna 'embark_town' e adiciona as colunas 'Queenstown' e 'Southampton' no DataFrame 'titanic'

titanic[['Queenstown', 'Southampton']] = pd.get_dummies(titanic.embark_town).drop(['Cherbourg'], axis=1)  

# O método 'get_dummies' cria variáveis dummy para todas as cidades de embarque, 
# e a função 'drop' remove a coluna 'Cherbourg', mantendo apenas as colunas 'Queenstown' e 'Southampton'
```

```python
# Remove a coluna 'embark_town' do DataFrame 'titanic'
titanic.drop(['embark_town'], axis=1, inplace=True)  # O parâmetro 'axis=1' indica que a operação é para colunas, e 'inplace=True' aplica a mudança diretamente no DataFrame original
```
