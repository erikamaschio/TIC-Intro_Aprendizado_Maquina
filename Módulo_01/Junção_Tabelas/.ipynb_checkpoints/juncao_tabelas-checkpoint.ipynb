```python
import pandas as pd
```

```python
voos = pd.read_csv('../../data/aulas/juncao_tabelas/voos.csv') #Lendo o arquivo
voos.head() #Visualizando as cinco primeiras linhas
```

```python
qnt_voos = pd.read_csv('../../data/aulas/juncao_tabelas/quatidade_de_voos.csv') #Lendo o arquivo
qnt_voos.tail() #Visualizando as cinco ultimas linhas
```

```python
pd.merge(voos, qnt_voos, on='month') #Merge, tenta encaixar uma tabela na outra, 'on=' indica onde precisa juntar
```

```python
produtos1 = pd.read_csv('../../data/aulas/juncao_tabelas/produtos1.csv', index_col=['id']) #Lendo o arquivo e trocando o id
produtos1.head() #Visualizando as cinco primeiras linhas
```

```python
produtos2 = pd.read_csv('../../data/aulas/juncao_tabelas/produtos2.csv', index_col=['id']) #Lendo o arquivo e trocando o id
produtos2.head() #Visualizando as cinco primeiras linhas
```

```python
produtos = pd.concat([produtos1, produtos2]) #Concatenando uma tabela encima da outra
produtos.head() #Visualizando as cinco primeiras linhas
```

```python
vendas = pd.read_csv('../../data/aulas/juncao_tabelas/vendas.csv') #Lendo o arquivo
vendas.head() #Visualizando as cinco primeiras linhas
```

```python
vendas = vendas.join(produtos, on='produto_id') #Agrupar as tabelas, 'on=' passar qual coluna eles se igualam 
```

```python
vendas.head() #Visualizando as cinco primeiras linhas
```

```python
vendas['total'] = vendas.quantidade * vendas.preco #Criando uma nova coluna e multiplicando os valores de 'quantidade' e 'preco'
```
