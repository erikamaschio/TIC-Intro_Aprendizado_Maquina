```python
from sklearn.preprocessing import MinMaxScaler
```

```python
iris.columns #Exibe as colunas
```

```python
# Define uma lista com os nomes das colunas numéricas do DataFrame 'iris'
numerical_columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
```

```python
# Exibe as primeiras 5 linhas das colunas numéricas do DataFrame 'iris'
iris[numerical_columns].head()
```

```python
# Cria uma instância do escalador MinMaxScaler, que normaliza os dados para um intervalo entre 0 e 1
scaler = MinMaxScaler()  

# Aplica o MinMaxScaler aos dados numéricos do DataFrame 'iris', ajustando e transformando os valores das colunas numéricas
scaled_data = scaler.fit_transform(iris[numerical_columns])  
```

```python
scaled_data
```

```python
# Converte os dados escalados de volta para um DataFrame, usando as colunas numéricas originais como nomes das colunas
scaled_data = pd.DataFrame(scaled_data, columns=numerical_columns)  
# Exibe as primeiras 5 linhas do DataFrame com os dados escalados
scaled_data.head()  
```
