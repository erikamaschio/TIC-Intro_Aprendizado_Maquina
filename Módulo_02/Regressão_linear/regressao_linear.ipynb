```python
# X_columns: Lista das variáveis independentes (features) que serão usadas para prever a qualidade do vinho.
X_columns = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol']

# y_columns: Lista com a variável dependente (alvo), representando a qualidade do vinho.
y_columns = ['quality'] 
```

```python
# X: DataFrame contendo as variáveis independentes (features) selecionadas.
X = wines[X_columns]

# y: DataFrame contendo a variável dependente (alvo), que é a qualidade do vinho.
y = wines[y_columns]
```

```python
X.head()
```

```python
y.head()
```

```python
# Importa a função train_test_split, que é usada para dividir os dados em conjuntos de treino e teste.
from sklearn.model_selection import train_test_split
```

```python
# Divide os dados em conjuntos de treino (70%) e teste (30%). 
# O parâmetro random_state=101 garante reprodutibilidade na divisão.

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
# Importa a classe LinearRegression, usada para criar e treinar modelos de regressão linear.
from sklearn.linear_model import LinearRegression
```

```python
# Cria uma instância do modelo de regressão linear.
lr = LinearRegression()
```

```python
# Treina o modelo de regressão linear usando os dados de treino (X_train e y_train).
lr.fit(X_train, y_train)
```

```python
# Retorna os coeficientes do modelo de regressão linear, que indicam a influência de cada feature na variável alvo.
lr.coef_
```

```python
# Faz previsões usando o modelo treinado, aplicando-o ao conjunto de teste (X_test).
preds = lr.predict(X_test)
```

```python
preds
```

```python
y_test
```

```python
# Cria um gráfico de dispersão (scatter plot) para comparar os valores reais (y_test) com os valores preditos (preds).
# A ejeção x exibe os valores reais, enquanto a ejeção y exibe os valores preditos pelo modelo.
# O gráfico ajuda a visualizar o desempenho do modelo de regressão.

plt.scatter(y_test, preds)
plt.xlabel("valores esperados")
plt.ylabel("valores preditos")
```

```python
# Importa a função mean_squared_error, que é usada para calcular o erro quadrático médio (MSE) entre os valores reais e preditos.
from sklearn.metrics import mean_squared_error
```

```python
# Calcula e exibe o erro quadrático médio (MSE) entre os valores reais (y_test) e os valores preditos (preds), que mede a precisão do modelo de regressão linear

print(mean_squared_error(y_test, preds))
```
