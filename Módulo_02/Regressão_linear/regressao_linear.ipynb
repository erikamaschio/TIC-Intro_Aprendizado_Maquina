```python
# X_columns: Lista das variveis independentes (features) que sero usadas para prever a qualidade do vinho.
X_columns = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol']

# y_columns: Lista com a varivel dependente (alvo), representando a qualidade do vinho.
y_columns = ['quality'] 
```

```python
# X: DataFrame contendo as variveis independentes (features) selecionadas.
X = wines[X_columns]

# y: DataFrame contendo a varivel dependente (alvo), que  a qualidade do vinho.
y = wines[y_columns]
```

```python
X.head()
```

```python
y.head()
```

```python
# Importa a funo train_test_split, que  usada para dividir os dados em conjuntos de treino e teste.
from sklearn.model_selection import train_test_split
```

```python
# Divide os dados em conjuntos de treino (70%) e teste (30%). 
# O parmetro random_state=101 garante reprodutibilidade na diviso.

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
# Importa a classe LinearRegression, usada para criar e treinar modelos de regresso linear.
from sklearn.linear_model import LinearRegression
```

```python
# Cria uma instncia do modelo de regresso linear.
lr = LinearRegression()
```

```python
# Treina o modelo de regresso linear usando os dados de treino (X_train e y_train).
lr.fit(X_train, y_train)
```

```python
# Retorna os coeficientes do modelo de regresso linear, que indicam a influncia de cada feature na varivel alvo.
lr.coef_
```

```python
# Faz previses usando o modelo treinado, aplicando-o ao conjunto de teste (X_test).
preds = lr.predict(X_test)
```

```python
preds
```

```python
y_test
```

```python
# Cria um grfico de disperso (scatter plot) para comparar os valores reais (y_test) com os valores preditos (preds).
# A ejeo x exibe os valores reais, enquanto a ejeo y exibe os valores preditos pelo modelo.
# O grfico ajuda a visualizar o desempenho do modelo de regresso.

plt.scatter(y_test, preds)
plt.xlabel("valores esperados")
plt.ylabel("valores preditos")
```

```python
# Importa a funo mean_squared_error, que  usada para calcular o erro quadrtico mdio (MSE) entre os valores reais e preditos.
from sklearn.metrics import mean_squared_error
```

```python
# Calcula e exibe o erro quadrtico mdio (MSE) entre os valores reais (y_test) e os valores preditos (preds), que mede a preciso do modelo de regresso linear

print(mean_squared_error(y_test, preds))
```
