```python
import pandas as pd
import seaborn as sns
```

```python
abalone = pd.read_csv('../../data/aulas/abalone.csv')
```

```python
# X_columns: Lista das variveis independentes (features) que representam as caractersticas fsicas dos moluscos, usadas para prever o sexo (Sex).
X_columns = ['Length', 'Diameter', 'Height', 'Whole weight', 'Shucked weight',
       'Viscera weight', 'Shell weight', 'Rings']

# y_columns: Lista contendo a varivel dependente (alvo), que  o sexo do molusco.
y_columns = ['Sex']
```

```python
X = abalone[X_columns]
y = abalone[y_columns]
```

```python
# Importa a funo train_test_split, que  usada para dividir os dados em conjuntos de treino e teste.
from sklearn.model_selection import train_test_split
```

```python
# Divide os dados em conjuntos de treino (70%) e teste (30%), utilizando um valor fixo para o random_state 
# para garantir que a diviso seja reprodutvel.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
# Exibe as dimenses do conjunto de treino (X_train), mostrando o nmero de amostras e caractersticas.
X_train.shape
```

```python
# Exibe as dimenses do conjunto de teste (X_test), mostrando o nmero de amostras e caractersticas.
X_test.shape
```

```python
# Importa a classe RandomForestClassifier, que  usada para criar um modelo de classificao baseado no algoritmo de floresta aleatria (Random Forest).
from sklearn.ensemble import RandomForestClassifier
```

```python
# Cria uma instncia do modelo RandomForestClassifier, configurado para usar 1000 rvores (n_estimators=1000) na floresta aleatria.
rf = RandomForestClassifier(n_estimators=1000)
```

```python
# Treina o modelo Random Forest usando os dados de treino (X_train e y_train).
rf.fit(X_train, y_train)
```

```python
# Faz previses usando o modelo Random Forest treinado, aplicando-o ao conjunto de teste (X_test).
preds = rf.predict(X_test)
```

```python
preds
```

```python
y_test
```

```python
# Importa a funo accuracy_score, que  usada para calcular a preciso do modelo, ou seja, a proporo de previses corretas.
from sklearn.metrics import accuracy_score
```

```python
# Calcula e exibe a preciso do modelo Random Forest, comparando os valores reais (y_test) com os valores preditos (preds).
# A preciso  a proporo de previses corretas feitas pelo modelo.
print(accuracy_score(y_test, preds))
```
